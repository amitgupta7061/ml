```python
import json
```

```python
# load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
        
    return data
```

```python
load_data("store_data.json")
```

```python
#Clean and structure data
def clean_data(data):
    text_tonum = {"one" : 1, "two" : 2, "three" : 3, "four" : 4, "five" : 5}
    cleaned_data = []
    unique_users = set()
    
    
    for user in data:
        # clean data
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_tonum):
            raw_rating = text_tonum[raw_rating]
        user["rating"] = raw_rating

        # missing info
        raw_age = user.get("age")
        if(raw_age == None):
            raw_age = None

        if(user["name"].strip() in unique_users):
            continue
        
        cleaned_data.append(user)
        unique_users.add(user["name"])
        
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
def get_insights(data):

    total_rating = 0
    for user in data:
        total_rating += float(user["rating"])

    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1

    print(f"avg rating = {total_rating / len(data)}")
    print(f"poor rating = {poor_rating / len(data) * 100}%")
```

```python
get_insights(data)
```

```python
# recommendation system like based on rating
def get_recommendation(data):
    recommendation = []

    for user in data:
        curr_recc = {}
        curr_recc["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_recc["brand"] = "apple"
        else:
            curr_recc["brand"] = "samsung"
        recommendation.append(curr_recc)

    return recommendation
            
```

```python
get_recommendation(data)
```
