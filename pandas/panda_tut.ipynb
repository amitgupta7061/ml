```python
# Cleaning 

df = pd.read_csv("raw_data.csv")
```

```python
df
df.isnull()
df.isna()

df.isnull().sum()  # giv esum of all null col

df.dropna()  # drop all the null value 
df.dropna(axis=1) # col drop

df.fillna("nhi")
df.fillna(df["age"].mean())

df.ffill() # forward fill
df.bfill() # backward fill 
```
