```python
import requests

URL = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(URL)
```

```python
print(response.status_code)
print(response.text)
print(response.headers)
print(response.content)
```

```python
with open("scrapped_data/data1.html", "w") as file:
    file.write(response.text)
```

```python
from bs4 import BeautifulSoup

with open("scrapped_data/data1.html", "r") as file:
    content = file.read()
soup = BeautifulSoup(content, "lxml")
# print(soup.prettify())
```

```python
import pandas as pd
soup.find("h1")
allH3 = soup.find_all("h3")

countryDetails = []

for h3 in allH3:
    country = h3.get_text(strip=True)
    # print(country.find_parent("div"))
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    countryDetails.append([country, population])


df = pd.DataFrame(countryDetails, columns=["Country", "Population"])
print(df)
```

```python
df.to_csv("scrapped_data/country_population.csv", index=False)
```
